## What is Flask Framework? What are the advantages of Flask Framework?
#### --> Flask is a Python web framework that allows developers to quickly and easily build web applications. It is lightweight, modular, and easy to use, making it a popular choice for developers who want to create small to medium-sized web applications. 

### Some advantages of Flask framework are:

#### 1] Lightweight: Flask is a lightweight framework that has few dependencies and does not require any particular tools or libraries. This makes it easy to install and get started with.

#### 2] Flexibility: Flask is a highly flexible framework that can be customized to meet the needs of any web application. 

#### 3] Easy to use: Flask is easy to use and does not require developers to have extensive knowledge of web development or programming. Its simple API and intuitive design make it easy for beginners to get started with.

#### 4] Scalability: Flask is highly scalable and can be used to build applications of any size. It provides developers with tools and libraries that can be used to optimize performance and ensure that applications can handle a large number of users and requests.

#### 5] Compatibility: Flask is compatible with a wide range of web servers and platforms, making it easy to deploy applications on different systems. 

#### 6] Large Community:Flask has a large community of developers who contribute to its development and support. This means that you can find a range of resources, including documentation, tutorials, and code examples, to help you learn and use the framework.

## What is App routing in Flask? Why do we use app routes?

#### --> In Flask, app routing refers to the process of defining the URL patterns (routes) that an application should respond to. It involves mapping a URL to a specific function that should be executed when the URL is accessed. This is done using the " @app.route "decorator in Flask.

#### --> App routes provide a flexible way to organize the functionality of an application into different URL patterns, allowing users to easily navigate between different pages and access different features. 

#### --> In addition, app routes make it easy to handle user input, as different routes can be used to process different types of input and generate appropriate responses.

## What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.
#### --> The url_for() function is used in Flask for URL building. It generates a URL based on the name of a function that handles the request for a particular route. The function takes the name of the function as its first argument, and any additional arguments can be passed to the function to generate a URL that includes those arguments.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return 'Home page'

@app.route('/hello/<name>')
def hello(name):
    return f'Hello, {name}!'

with app.test_request_context():
    print(url_for('index'))
    print(url_for('hello', name='Harsh'))
```
