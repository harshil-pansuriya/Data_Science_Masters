# 4.
## What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.

#### --> In Pandas, a DataFrame is a two-dimensional labeled data structure, consisting of rows and columns, similar to a spreadsheet or a SQL table.
#### --> On the other hand, a Series is a one-dimensional labeled array capable of holding any data type (integers, strings, floating point numbers, Python objects, etc.).

```python
import pandas as pd 
```

```python
df3= pd.Series([1,2,3,4])
df3
```

```python
df4 = pd.DataFrame({'col1': [1, 2, 3], 'col2': ['a', 'b', 'c']}, index=['A', 'B', 'C'])
df4
```
