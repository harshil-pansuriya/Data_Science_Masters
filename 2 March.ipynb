## What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.
#### --> Matplotlib is used for data visualization in a wide range of fields, including science, engineering, finance, and economics. It can be used to create a wide variety of plots and charts, including scatter plots, line plots, bar plots, histograms, and pie charts.

### Here are five plots that can be created using the Pyplot module of Matplotlib:

#### 1] Scatter plot: A scatter plot is a plot where the values of two variables are plotted along the X and Y axis, and the points on the plot represent the values of the two variables for each data point.
#### 2] Line plot: A line plot is a plot where the values of one variable are plotted against time or another continuous variable, and the line connecting the points represents the trend in the data.
#### 3] Bar plot: A bar plot is a plot where the values of one variable are represented using bars of varying heights.
#### 4] Histogram: A histogram is a plot that represents the distribution of a single variable. The X-axis represents the variable and the Y-axis represents the frequency or count of the values.
#### 5] Pie chart: A pie chart is a chart that represents the proportion of different categories in a dataset using slices of a circle.

# 3.
## What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
#### --> A scatter plot is a graph that displays the relationship between two quantitative variables. It is a plot where the values of two variables are plotted along the X and Y axis, and the points on the plot represent the values of the two variables for each data point.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y, c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.grid(True)
```

# 3.
## Why is the subplot() function used? Draw four line plots using the subplot() function.
#### --> The subplot() function in Matplotlib is used to create multiple plots in the same figure. It allows you to divide a figure into a grid of subplots and plot different data in each subplot. This function is useful when you want to compare different plots or display multiple plots in the same figure.

```python
import numpy as np
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])

fig,axs=plt.subplots(4)
axs[0].plot(x1,y1,c="b")
axs[1].plot(x2,y2,c="g")
axs[2].plot(x3,y3,c="r")
axs[3].plot(x4,y4,c="y")
plt.show()
```

#### --> A box plot, also known as a box-and-whisker plot, is a type of chart used to display the distribution of a dataset. It summarizes the five-number summary of a dataset, which includes the minimum, maximum, median, and first and third quartiles. Box plots are useful for comparing the distribution of different datasets and identifying outliers.

```python
box1 = np.random.normal(100, 10, 200)
plt.boxplot(box1)
plt.show()
```
